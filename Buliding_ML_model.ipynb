# **Load Data**

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv');
df
```

## **Data Preparation**

## **Data Separation**

```python
y = df['logS']
y
```

```python
X = df.drop('logS', axis=1)
X
```

##**Data Splitting**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=200)
```

# **Model Building**

## **Linear Regression**

### **Training the model**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

### **Applying the model to make prediction**

```python
y_lr_train_pred = lr.predict(X_train)
y_lr_test_pred = lr.predict(X_test)

print(y_lr_train_pred)
```

```python
print(y_lr_test_pred)
```

### **Evaluate Model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_rs = r2_score(y_train, y_lr_train_pred)

lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_rs = r2_score(y_test, y_lr_test_pred)
```

```python
lr_train_mse
```

```python
lr_train_rs
```

```python
lr_test_mse
```

```python
lr_test_rs
```

```python
lr_results = pd.DataFrame(['Linear Regression',lr_train_mse, lr_train_rs, lr_test_mse, lr_test_rs]).transpose();
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
lr_results
```

## **Random Forest**

### **Training the model**

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor(max_depth=2, random_state=100)
rf.fit(X_train, y_train)
```

### **Applying the model to make prediction**

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)

print(y_rf_train_pred)
```

```python
print(y_rf_test_pred)
```

### **Evaluate Model Performance**

```python
from sklearn.metrics import mean_squared_error, r2_score

rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_rs = r2_score(y_train, y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_rs = r2_score(y_test, y_rf_test_pred)
```

```python
print(rf_train_mse)
print(rf_train_rs)
print(rf_test_mse)
print(rf_test_rs)
```

```python
rf_results = pd.DataFrame(['Random Forest',rf_train_mse, rf_train_rs, rf_test_mse, rf_test_rs]).transpose();
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
rf_results
```

```python
df_models = pd.concat([lr_results, rf_results])
df_models
```

```python
df_models.reset_index(drop=True)
```
